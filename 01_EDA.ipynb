# SKU Cleaning

```python
sku = pd.read_csv('Dillards_POS/skuinfo.csv', delimiter='\t', header=None)
sku = pd.DataFrame(sku[0].str.split(',', expand=True))
sku
```

```python
bad_formatted_rows = sku[~sku[10].apply(lambda x: str(x).isdigit())]
bad_formatted_rows
```

```python
bad_formatted_rows[9] = bad_formatted_rows[9] + ',' + bad_formatted_rows[10]
bad_formatted_rows[9] = bad_formatted_rows[9].apply(lambda x: x.strip(', '))
bad_formatted_rows
```

```python
cleaned_sku = pd.concat([sku.drop(bad_formatted_rows.index), bad_formatted_rows])

# Reset the index of the merged DataFrame
cleaned_sku.reset_index(drop=True, inplace=True)

cleaned_sku
```

```python
cleaned_sku.drop(columns=[10], inplace=True)
cleaned_sku.columns = ['sku', 'dept', 'classid', 'upc', 'style', 'color', 'size', 
                       'packsize', 'vendor', 'brand', 'unknown', 'extra']
cleaned_sku
```

```python
cleaned_sku['sku'] = cleaned_sku['sku'].astype(int)
cleaned_sku['dept'] = cleaned_sku['dept'].astype(int)

cleaned_sku.dtypes
```

```python
d = set(deptinfo['deptID'].unique())
d
```

```python
s = set(cleaned_sku['dept'].unique())
s
```

```python
s == d
```

```python
# Find elements that are in s1 but not in set2
difference = d.difference(s)
print("Elements in deptinfo but not in sku:", difference)

# Find elements that are in set2 but not in set1
difference_reverse = s.difference(d)
print("Elements in sku but not in deptinfo:", difference_reverse)
```

```python
bad_rows = cleaned_sku[~cleaned_sku['packsize'].apply(lambda x: str(x).isdigit())|
    ~cleaned_sku['vendor'].apply(lambda x: str(x).isdigit())]
bad_rows
```

```python
# these im going to try to fix
first_set = bad_rows[bad_rows['unknown'].apply(lambda x: str(x).isdigit())]
first_set
```

```python
first_set['color'] = first_set['color'] + ', ' + first_set['size']
first_set['size'] = first_set['packsize']
first_set['packsize'] = first_set['vendor']
first_set[['vendor', 'brand']] = first_set['brand'].str.split(',', expand=True)
first_set['color'] = first_set['color'].apply(lambda x: x.strip(', '))
first_set
```

```python
# some sizes are still messed up, but shouldnt affect our analysis
check_first_set = first_set[~first_set['size'].apply(lambda x: str(x).strip().isdigit())]
check_first_set
```

```python
# these dont have a valid 'vendor', that should be fine as we wont be using vendor in our analysis
# no idea what CHTI or G means, just dropping those
second_set = bad_rows[~bad_rows['unknown'].apply(lambda x: str(x).isdigit())]
second_set
```

```python
second_set['size'] = second_set['vendor']
second_set['packsize'] = second_set['brand']
second_set['brand'] = second_set['unknown']
second_set['vendor'] = None
second_set['unknown'] = '0'
second_set['extra'] = None
second_set
```

```python
cleaned_sku = pd.concat([cleaned_sku.drop(first_set.index), first_set])
cleaned_sku = pd.concat([cleaned_sku.drop(second_set.index), second_set])
cleaned_sku = cleaned_sku.sort_values(['sku'])
cleaned_sku
```

```python
# check if we missed anything from the last 2 columns
cleaned_sku[~cleaned_sku['extra'].isna()]
```

```python
check = cleaned_sku[~cleaned_sku['unknown'].isna()]
check[~check['unknown'].apply(lambda x: str(x).isdigit())]
```

```python
# check if columns have correct data types
# packsize and vendor should be integers
cleaned_sku[~cleaned_sku['packsize'].apply(lambda x: str(x).isdigit())]
```

```python
cleaned_sku[~cleaned_sku['vendor'].apply(lambda x: str(x).isdigit())]
```

```python
# change data types, changed nan vendors to 0
cleaned_sku['vendor'] = cleaned_sku['vendor'].fillna(0).astype(int)

cleaned_sku['packsize'] = cleaned_sku['packsize'].astype(int)

cleaned_sku.dtypes
```

```python
# now figure out the classid before exporting
bad_classid = cleaned_sku[~cleaned_sku['classid'].apply(lambda x: str(x).isdigit())]
bad_classid.head(50)
```
