```python
trans[trans['amt'] != trans['amt_dupe']]
trans.drop('amt_dupe', axis = 1, inplace = True)
```

```python
trans.head()
```

```python
trans['orgprice'].replace(0.0, np.nan, inplace=True)
```

```python
trans['orgprice'] = trans.groupby('sku')['orgprice'].fillna(method='ffill')
```
